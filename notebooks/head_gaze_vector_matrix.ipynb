```python
import h5py
import matplotlib.pyplot as plt
def plot_face_gaze(ax, h5_path, title, cam_index=None, label=None):
    with h5py.File(h5_path, 'r', swmr=True) as source_f:
        face_gaze = source_f['face_gaze']
        yaw = face_gaze[:,1]*180/np.pi
        pitch = face_gaze[:,0]*180/np.pi
        
        if cam_index is not None:
            cam_index_data = source_f['cam_index'][:]
            cam_index_selected = np.isin(cam_index_data, cam_index)
            yaw = yaw[cam_index_selected]
            pitch = pitch[cam_index_selected]

    
    

    color_dict = {'A': 'red', 'B': 'green', 'C': 'blue'}

    # Map the labels to colors using the dictionary
    if label is not None:
        colors = color_dict[label]
    else:
        colors = 'blue' # default color if no labels are given

    # Plot pitch and yaw angles with colors and transparency
    ax.scatter(yaw, pitch, c=colors, alpha=0.05)
```

```python

# Create a square graph
for cam in range(1,19):
    # Set x and y limits
    fig, ax = plt.subplots(figsize=(4,4))
    ax.set_xlim([-130,130])
    ax.set_ylim([-130,130])

    # Set axis labels
    ax.set_xlabel('Yaw')
    ax.set_ylabel('Pitch')

    # Add a grid
    ax.grid(True)
    ax.set_title(f'Cam{cam}')

    plot_face_gaze(ax, "/home/jqin/Datasets/MPIIFaceGaze_normalized_h5/p00.h5", 'mpii')
    plot_face_gaze(ax, "/work/jqin/diffusion_iccv/xgaze_512/train/subject0085.h5", 'Xgaze cam 1 2',
                   cam_index = [cam], label='A')

    plt.show()
```
